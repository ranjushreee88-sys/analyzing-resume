```python
import pandas as pd
import spacy

nlp = spacy.load("en_core_web_sm")
```

```python
import pandas as pd

df = pd.read_csv("/content/Resume.csv", engine='python', on_bad_lines='skip')

df.head()
```

```python
def extract_entities(text):
  doc = nlp(text)

  names = []
  orgs = [] # Changed 'args' to 'orgs' for consistency

  for ent in doc.ents:
    if ent.label_ == "PERSON":
      names.append(ent.text)
    elif ent.label_ == "ORG": # Separate condition for organizations
      orgs.append(ent.text)

  return names, orgs # Return both lists after processing all entities
```

```python
for text in df['Resume_str'].head(5):

   names, orgs = extract_entities(text)

   print("Names:", names)
   print("Organization:", orgs)
   print("---------------------")
```
